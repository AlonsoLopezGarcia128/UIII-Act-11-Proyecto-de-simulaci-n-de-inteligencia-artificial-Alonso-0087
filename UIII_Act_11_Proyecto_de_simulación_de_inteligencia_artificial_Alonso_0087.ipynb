```python
# --- SIMULADOR DE IA: GESTIÓN DE FOTÓGRAFOS ---

fotografos = []

# ---------------- FUNCIONES ----------------

def mostrar_menu():
    print("\n===== MENÚ PRINCIPAL =====")
    print("1. Agregar fotógrafo")
    print("2. Mostrar fotógrafos (tabla)")
    print("3. Buscar fotógrafo por especialidad")
    print("4. Salir")

def agregar_fotografo():
    print("\n--- REGISTRO DE FOTÓGRAFO ---")

    nombre = input("Nombre del fotógrafo: ")
    camara = input("Tipo de cámara: ")
    lentes = input("Lentes disponibles: ")
    especialidad = input("Especialidad (bodas/moda): ")
    estudio = input("Estudio: ")
    portafolio = input("Portafolio: ")
    resolucion = input("Resolución de salida: ")

    fotografo = {
        "Nombre": nombre,
        "Tipo de cámara": camara,
        "Lentes disponibles": lentes,
        "Especialidad": especialidad,
        "Estudio": estudio,
        "Portafolio": portafolio,
        "Resolución": resolucion
    }

    fotografos.append(fotografo)
    print("✅ Fotógrafo agregado correctamente")

def mostrar_fotografos():
    print("\n--- LISTA DE FOTÓGRAFOS ---")

    if len(fotografos) == 0:
        print("⚠ No hay fotógrafos registrados")
    else:
        # Encabezado
        print("+----------------------+----------------------+----------------------+----------------------+----------------------+----------------------+----------------------+")
        print("| {:<20} | {:<20} | {:<20} | {:<20} | {:<20} | {:<20} | {:<20} |".format(
            "Nombre", "Cámara", "Lentes", "Especialidad", "Estudio", "Portafolio", "Resolución"
        ))
        print("+----------------------+----------------------+----------------------+----------------------+----------------------+----------------------+----------------------+")

        # Filas
        for f in fotografos:
            print("| {:<20} | {:<20} | {:<20} | {:<20} | {:<20} | {:<20} | {:<20} |".format(
                f["Nombre"],
                f["Tipo de cámara"],
                f["Lentes disponibles"],
                f["Especialidad"],
                f["Estudio"],
                f["Portafolio"],
                f["Resolución"]
            ))
            print("+----------------------+----------------------+----------------------+----------------------+----------------------+----------------------+----------------------+")

def buscar_por_especialidad():
    print("\n--- BÚSQUEDA ---")
    esp = input("Ingresa especialidad: ")

    encontrados = 0

    for f in fotografos:
        if f["Especialidad"] == esp:
          print("\n🎯 Coincidencia encontrada:")
          print("Nombre:", f["Nombre"])
          print("Cámara:", f["Tipo de cámara"])
          print("Lentes:", f["Lentes disponibles"])
          print("Estudio:", f["Estudio"])
          encontrados += 1

    if encontrados == 0:
        print("❌ No se encontraron resultados")

# ---------------- PROGRAMA PRINCIPAL ----------------

print("🤖 INICIALIZANDO IA DE FOTÓGRAFOS 2026...")

while True:
    mostrar_menu()
    opcion = input("Selecciona una opción: ")

    if opcion == "1":
        agregar_fotografo()

    elif opcion == "2":
        mostrar_fotografos()

    elif opcion == "3":
        buscar_por_especialidad()

    elif opcion == "4":
        print("👋 Saliendo del sistema...")
        break

    else:
        print("⚠ Opción no válida")
```
